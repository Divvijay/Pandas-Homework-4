# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
csvpath_whale_returns = Path("../Resources/whale_returns.csv")
```

```python
# Reading whale returns
whale_returns_df =pd.read_csv(csvpath_whale_returns,index_col ="Date",infer_datetime_format =True,parse_dates =True)
whale_returns_df
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
#Reading algorithmic returns
csvpath_Algo_returns = Path("../Resources/algo_returns.csv")
Algo_returns_df =pd.read_csv(csvpath_Algo_returns,index_col ="Date",infer_datetime_format =True,parse_dates =True)
Algo_returns_df
```

```python
# Count nulls
Algo_returns_df.isnull().count()
```

```python
# Drop nulls
Algo_returns_df.dropna(inplace = True)
Algo_returns_df
```

```python
#Checking for if the Nulss are removed. Since the outcome is 0 it implies that the null values are removed.
Algo_returns_df.isnull().mean()*100
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
csvpath_sp_tsx_history = Path("../Resources/sp_tsx_history.csv")
sp_tsx_history_df =pd.read_csv(csvpath_sp_tsx_history,index_col ="Date",infer_datetime_format =True,parse_dates =True)
sp_tsx_history_df
```

```python
# Check Data Types
sp_tsx_history_df.dtypes
```

```python
# Fix Data Types - removing the dollar sign and , from the numbers
sp_tsx_history_df["Close"] = sp_tsx_history_df["Close"].str.replace("$","")
sp_tsx_history_df["Close"] = sp_tsx_history_df["Close"].str.replace(",","")
sp_tsx_history_df
```

```python
# Converting the "Close" column from String to Float
sp_tsx_history_df["Close"]= sp_tsx_history_df["Close"].astype("float")
sp_tsx_history_df
```

```python
# Calculate Daily Returns
sp_tsx_daily_returns_df = sp_tsx_history_df.pct_change()
sp_tsx_daily_returns_df
```

```python
# Drop nulls
sp_tsx_daily_returns_df.dropna(inplace = True)
sp_tsx_daily_returns_df
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_tsx_daily_returns_df.rename(columns={"Close": "TSX Returns"})
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
Combined_Portfolio_daily_returns = pd.concat([whale_returns_df,Algo_returns_df,sp_tsx_daily_returns_df],axis = "columns",join = "inner")
Combined_Portfolio_daily_returns
```

```python
#Renaming the "Close" column in the Combined Portfolio Data Frame.
Combined_Portfolio_daily_returns.rename(columns={"Close": "TSX Returns"},inplace=True)
Combined_Portfolio_daily_returns
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&P TSX 60 using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window

Combined_Portfolio_Rolling_std = Combined_Portfolio_daily_returns.rolling(window = 21).std()

# Plot the rolling standard deviation
Combined_Portfolio_Rolling_std.plot(figsize = (25,12),title = "21 Day Rolling Standard Deviation of all Portfolios and TSX")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
Combined_Portfolio_Correlation = Combined_Portfolio_daily_returns.corr()
# Display de correlation matrix
Combined_Portfolio_Correlation
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
Cov_Soros_Fund = Combined_Portfolio_daily_returns["SOROS FUND MANAGEMENT LLC"].cov(Combined_Portfolio_daily_returns["TSX Returns"])
Cov_Paulson_Fund = Combined_Portfolio_daily_returns["PAULSON & CO.INC. "].cov(Combined_Portfolio_daily_returns["TSX Returns"])
Cov_Tiger_Global = Combined_Portfolio_daily_returns["TIGER GLOBAL MANAGEMENT LLC"].cov(Combined_Portfolio_daily_returns["TSX Returns"])
Cov_Bershire_Hathway = Combined_Portfolio_daily_returns["BERKSHIRE HATHAWAY INC"].cov(Combined_Portfolio_daily_returns["TSX Returns"])
Cov_Algo_1 = Combined_Portfolio_daily_returns["Algo 1"].cov(Combined_Portfolio_daily_returns["TSX Returns"])                                                                                         
Cov_Algo_2 = Combined_Portfolio_daily_returns["Algo 2"].cov(Combined_Portfolio_daily_returns["TSX Returns"])
                                                                                                                        
# Calculate variance of S&P TSX

variance_TSX = Combined_Portfolio_daily_returns["TSX Returns"].var()

```

```python

print(f"Covariance of Soros Fund:{Cov_Soros_Fund}\n Covriance Paulson & Co.:{Cov_Paulson_Fund}\n Covariance Tiger Management LLC:{Cov_Tiger_Global}\n Covariance BERKSHIRE HATHAWAY INC:{Cov_Bershire_Hathway}\n Covariance Algo 1:{Cov_Algo_1}\n Covariance Algo 2:{Cov_Algo_2}")
```

```python
# Computing betas

Beta_Soros_Fund = (Cov_Soros_Fund /variance_TSX)
Beta_Paulson_Fund = (Cov_Paulson_Fund/variance_TSX)
Beta_Tiger_Global = (Cov_Tiger_Global/variance_TSX)
Beta_Bershire_Hathway = (Cov_Bershire_Hathway/variance_TSX)
Beta_Algo_1 = (Cov_Algo_1/variance_TSX)
Beta_Algo_2 = (Cov_Algo_2/variance_TSX)

print("Betas of all funds are as follows:")
print(f" Soros Fund Beta: {Beta_Soros_Fund}\n Paulson Fund Beta: {Beta_Paulson_Fund } \n Tiger Global Management Beta : {Beta_Tiger_Global}\n Algo1 Beta :{Beta_Algo_1}\n Algo 2 Beta: {Beta_Algo_2}")
```

```python
#21 Days rolling betas of all Portfolios

#Step 1: Calculating the 21 Day Rolling Covriances of all Portfolios
Cov_21Day_Rolling_Soros_Fund = Combined_Portfolio_daily_returns["SOROS FUND MANAGEMENT LLC"].rolling(window=21).cov(Combined_Portfolio_daily_returns["TSX Returns"])
Cov_21Day_Rolling_Paulson_Fund = Combined_Portfolio_daily_returns["PAULSON & CO.INC. "].rolling(window=21).cov(Combined_Portfolio_daily_returns["TSX Returns"])
Cov_21Day_Rolling_Tiger_Global = Combined_Portfolio_daily_returns["TIGER GLOBAL MANAGEMENT LLC"].rolling(window=21).cov(Combined_Portfolio_daily_returns["TSX Returns"])
Cov_21Day_Rolling_Bershire_Hathway = Combined_Portfolio_daily_returns["BERKSHIRE HATHAWAY INC"].rolling(window=21).cov(Combined_Portfolio_daily_returns["TSX Returns"])
Cov_21Day_Rolling_Algo_1 = Combined_Portfolio_daily_returns["Algo 1"].rolling(window=21).cov(Combined_Portfolio_daily_returns["TSX Returns"])                                                                                         
Cov_21Day_Rolling_Algo_2 = Combined_Portfolio_daily_returns["Algo 2"].rolling(window=21).cov(Combined_Portfolio_daily_returns["TSX Returns"])

#Step 2: Calculating the 21 Day Rolling Variance of the TSX

Rolling_21days_variance_TSX = Combined_Portfolio_daily_returns["TSX Returns"].rolling(window=21).var()

#Step3: Calculating the 21 Day Rolling Betas of all Portfolios 

Rolling_21Days_Beta_Soros_Fund = (Cov_21Day_Rolling_Soros_Fund /Rolling_21days_variance_TSX )
Rolling_21Days_Beta_Paulson_Fund = (Cov_21Day_Rolling_Paulson_Fund /Rolling_21days_variance_TSX )
Rolling_21Days_Beta_Tiger_Global = (Cov_21Day_Rolling_Tiger_Global /Rolling_21days_variance_TSX )
Rolling_21Days_Beta_Bershire_Hathway = (Cov_21Day_Rolling_Bershire_Hathway /Rolling_21days_variance_TSX )
Rolling_21Days_Beta_Algo_1 = (Cov_21Day_Rolling_Algo_1 /Rolling_21days_variance_TSX )
Rolling_21Days_Beta_Algo_2 = (Cov_21Day_Rolling_Algo_2 /Rolling_21days_variance_TSX )
```

```python
#Plot of 21 Days Rolling Betas
ax = Rolling_21Days_Beta_Soros_Fund.plot(figsize=(25,12),title="21 Day Rolling Betas of all Portfolios")
Rolling_21Days_Beta_Paulson_Fund.plot(ax =ax)
Rolling_21Days_Beta_Tiger_Global.plot(ax =ax)
Rolling_21Days_Beta_Bershire_Hathway.plot(ax =ax)
Rolling_21Days_Beta_Algo_1.plot(ax =ax)
Rolling_21Days_Beta_Algo_2.plot(ax =ax)
ax.legend(["Soros Fund","Paulson Fund","Tiger Global","Berkshire Hathway","Algo1","Algo2"])
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
# calculating the exponentially weighted moving average of the daily returns of all portfolios.
Combined_Portfolio_daily_returns_EWM = Combined_Portfolio_daily_returns.ewm(halflife =21).mean()
Combined_Portfolio_daily_returns_EWM
```

```python
#21 Day halflife ewm - average returns
Combined_Portfolio_daily_returns_EWM.plot(figsize = (25,12))
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios

#Step 1: Calculate the annulaized Mean of all stocks
Combined_Portfolio_annualized_mean = Combined_Portfolio_daily_returns.mean()*252
#Step 2: Calculate the Sharpe Ratio using the formule = (annualized mean - Risk Free Return)/annualized Standard Deviation
# I have assumed the risk free rate to be 0%.
Combined_Portfolio_Sharpe_Ratios = (Combined_Portfolio_annualized_mean) /(Combined_Portfolio_Annualized_std)
Combined_Portfolio_Sharpe_Ratios.sort_values(inplace = True)
Combined_Portfolio_Sharpe_Ratios
```

```python
# Visualize the sharpe ratios as a bar plot
Combined_Portfolio_Sharpe_Ratios.plot(kind ="bar",title ="Sharpe Ratios of all Portfolios")
```

### Determine whether the algorithmic strategies outperform both the market (S&P TSX 60) and the whales portfolios.

The Sharpe Ratio is a measure of comparing the peroformance of portfolios on a risk adjusted return basis. It gives the return eanred per
unit of riks. So the higher the Sharpe Ratio the better it is. In this case the Algo 1 portfolio has the highest sharpe ratio and it beats the Whale Portfolio and the benchmark (TSX).Therefore we can say that Algo 1 performed the best as compared to the benchmark and all other portfolios.
Secondly, as far as the Algo 2 portfolio is concerned, it outperforms the TSX and other portfolios except Berkshire Hathway Inc.

```python
# Set weights

weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
My_Portfolio_daily_returns = My_Portfolio_all_stocks_daily_returns.dot(weights)
# Display sample data
My_Portfolio_daily_returns.head()
```
